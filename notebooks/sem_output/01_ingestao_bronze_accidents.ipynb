# Ingestão Camada Bronze

```python
%python
# Caminho
raw_path = "/Volumes/workspace/default/raw_accident/accidents_2017_to_2023_portugues.csv"

# Leitura do CSV
df_raw = (
    spark.read
    .option("header", True)
    .option("sep", ",")
    .option("inferSchema", True)
    .csv(raw_path)
)
display(df_raw)
```

```python
%sql
CREATE VOLUME IF NOT EXISTS workspace.default.bronze_accident;
```

```python
bronze_path = "/Volumes/workspace/default/bronze_accident"
(
    df_raw
    .write
    .format("delta")
    .mode("overwrite")
    .save(bronze_path)
)
```

```python
# Validação
df_raw_check = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(raw_path)
)
bronze_path = "/Volumes/workspace/default/bronze_accident"
df_bronze = spark.read.format("delta").load(bronze_path)
print("raw rows:", df_raw_check.count())
print("bronze rows:", df_bronze.count())
```
